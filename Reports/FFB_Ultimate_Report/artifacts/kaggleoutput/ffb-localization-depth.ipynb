```python
%%writefile ffb_localization_depth.yaml
path: /kaggle/input/ffb-localization-depth-dataset/ffb_localization_depth
train: images/train
val: images/val
test: images/test
nc: 1
names: ['fresh_fruit_bunch']
```

```python
from ultralytics import YOLO
model = YOLO("yolo11n.pt")

model.train(
    data="/kaggle/working/ffb_localization_depth.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device="cuda",
    seed=42,
    project="/kaggle/working/runs/detect",
    name="exp_a2_depth_seed42",
)

!zip -r /kaggle/working/exp_a2_depth_seed42.zip /kaggle/working/runs/detect/exp_a2_depth_seed42

YOLO("/kaggle/working/runs/detect/exp_a2_depth_seed42/weights/best.pt").val(
    data="ffb_localization_depth.yaml", split="test", project="/kaggle/working/runs/detect", name="test_seed42"
)
```
