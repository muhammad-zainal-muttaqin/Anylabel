```python
%%writefile ffb_localization.yaml
path: /kaggle/input/ffb-localization-dataset/ffb_localization
train: images/train
val: images/val
test: images/test
nc: 1
names: ['fresh_fruit_bunch']
```

```python
from ultralytics import YOLO
model = YOLO("yolo11n.pt")

model.train(
    data="/kaggle/working/ffb_localization.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device="cuda",
    seed=42,
    project="/kaggle/working/runs/detect",
    name="exp_a1_rgb_seed42",
)
```
